## Time Series visualisation

```python
for df in (ibm, msft, nvda):
    df['Open'].plot()
```

```python
def plot_timeseries(column):

    plt.figure(figsize=(12, 18))

    plt.subplot(3, 1, 1)
    plt.title(f'IBM - {column}')
    plt.plot(
        ibm[column]
    )

    plt.subplot(3, 1, 2)
    plt.title(f'MSFT - {column}')
    plt.plot(
        msft[column]
    )

    plt.subplot(3, 1, 3)
    plt.title(f'NVDA - {column}')
    plt.plot(
        nvda[column]
    )
    plt.show()
    pass
```

```python
plot_timeseries('Open')
```

```python
plot_timeseries('Close')
```

```python
plot_timeseries('High')
```

```python
plot_timeseries('Low')
```

```python
plot_timeseries('Close')
```

```python
plot_timeseries('Volume')
```

```python
plot_timeseries('Dividends')
```

```python
plot_timeseries('Stock Splits')
```

#### Notes
Visualising the series shows periods of (overall) increases and decreases, this could cause a problem with learning as the model may need good examples of both. This could potentially be avoided by keeping a models prediction window smaller. Experiments would need to be created to find the most optimal window size.

Another way of preventing this could be to ignore the temporal element of the dataset and focus only predicting based on past values.

To improve the chance getting an accurate model, the aim will be to predict if the next day will be an increase or a decrease. This is a variation on an existing type of algorithmic trading called 'high-frequency trading'. Although our model will not be trading with such high frequency I see no reason why this would not scale to day by day trades.

For this a target variable will need to be created from the data.

Details of this can be found in the `FeatureEngineeringDevelopment.ipynb` notebook, with functions saved in `src/dataPreparation`.

## Create target variable


```python
# Import preparation tools
import src.dataPreparation as prep

# Create target variables
ibm = prep.calculateTarget(ibm)
nvda = prep.calculateTarget(nvda)
msft = prep.calculateTarget(msft)

# Change target dtype to category
ibm['target'] = ibm['target'].astype('category')
nvda['target'] = nvda['target'].astype('category')
msft['target'] = msft['target'].astype('category')
```

## Correlation Matrices

```python
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
sns.heatmap(
    ibm.corr(numeric_only=True),
    vmin=-1, vmax=1,
)
plt.title('IBM Correlation Matrix')

plt.subplot(1, 3, 2)
sns.heatmap(
    msft.corr(numeric_only=True),
    vmin=-1, vmax=1
)
plt.title('MSFT Correlation Matrix')

plt.subplot(1, 3, 3)
sns.heatmap(
    nvda.corr(numeric_only=True),
    vmin=-1, vmax=1
)
plt.title('NVDA Correlation Matrix')
```

### Notes
Correlational matrices of all three datasets show there is a high level of cardinality. This suggests LSTM models maybe useful for modeling this dataset.
